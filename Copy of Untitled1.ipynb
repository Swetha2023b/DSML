```python
#dataframe from series
import pandas as pd
s = pd.Series(['a','b','c','d'])
df=pd.DataFrame(s)
print(df)
```

```python
#dataframe from list of dictionaries
import pandas as pd
d=[{'Name':'Swetha','Subject':'Astrophysics'},{'Name':'Susan','Subject':'Machine Learning'},{'Name':'Upas','Subject':'Photography'}]
df=pd.DataFrame(d)
print(df)
```
